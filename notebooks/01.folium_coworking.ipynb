```python
import folium
import json
```

```python
with open('./data/locations.json') as json_file:
    data = json.load(json_file)
```

```python
m = folium.Map(location=[49.6, 5.72], zoom_start=11)
```

```python
#data
```

```python
for coworking in data['coworking']:
    popup = f"<b>{coworking['name']}</b></br>"
    popup += f"{coworking['founder']}</br>"
    popup += f"{coworking['address']['street']} {coworking['address']['number']}</br>"
    popup += f"{coworking['address']['postalCode']} {coworking['address']['city']}</br>"
    popup += f'<a href="{coworking["website"]}">{coworking["website"]}</a></br>'
    popup += f'<a href="{coworking["mail"]}">{coworking["mail"]}</a></br>'
    folium.Marker(coworking['coordinates'], popup=popup, tooltip =coworking['name'] ).add_to(m)
    
```
